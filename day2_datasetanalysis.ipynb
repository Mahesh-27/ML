```python
# data preparation for modeling
# encoding => converting non-numerical data to numerical data
#train test split
```

```python
d.head()
```

```python
non_num=d.select_dtypes("object")#gives non-numerical data
non_num
num_col=d.select_dtypes("number")#gives numerical data
enc_col=pd.get_dummies(non_num)#one hot encoding
enc_col
#joining the encoded data with numerical data
d_new=pd.concat([num_col,enc_col],axis=1)#axis=1 for merging it as column and vice-versa.
d_new
```

```python
from sklearn.model_selection import train_test_split
```

```python
y=d_new['Item_Outlet_Sales']#target 
x=d_new.drop(columns="Item_Outlet_Sales")#independent varibale
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
#Data Modeling
from sklearn.linear_model import LinearRegression 
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.intercept_
```

```python
lr.coef_
```

```python
#industrial usage model
import statsmodels.api as sm
x_train_cons=sm.add_constant(x_train)#adding constant to get the intercept
lr_ols=sm.OLS(y_train,x_train_cons).fit()
lr_ols.summary()
```

```python
x_test_cons=sm.add_constant(x_test)
y_pred=lr_ols.predict(x_test_cons)
y_pred
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_pred,y_test)**0.5
```
